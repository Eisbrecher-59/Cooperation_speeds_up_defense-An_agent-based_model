```python
label = []
for i in range(1, subgrid + 1):
    if i % 2 == 1:
        label.append("")
    else:
        label.append(str(i))

os.chdir("/home/gillard/Bureau/Book_Project/Towards_a_response_time_of_zero/LaTeX Material/Results")
# scenario = os.listdir("/home/gillard/Bureau/Book_Project/Towards_a_response_time_of_zero/LaTeX Material/Results/Results_AB/grids_A")
# print(scenario)
scenario = ["iter-100_l-50_h-50_d_A-0.250_d_B-0.250_m_r_A-2_m_r_B-2_r_m_A-0.200_r_r_A-0.300_g_r_A-0.500_d_r_A-0.100_t_l_A-0.900_p-0.100_t-1.300_i_c_l_B-0.500_i_l_B-1.000_0.csv",
            "iter-100_l-50_h-50_d_A-0.250_d_B-0.250_m_r_A-2_m_r_B-2_r_m_A-0.200_r_r_A-0.500_g_r_A-0.100_d_r_A-0.500_t_l_A-0.900_p-0.100_t-1.300_i_c_l_B-0.500_i_l_B-1.000_0.csv", 
            "iter-100_l-50_h-50_d_A-0.250_d_B-0.250_m_r_A-2_m_r_B-2_r_m_A-0.200_r_r_A-0.200_g_r_A-0.200_d_r_A-0.100_t_l_A-0.900_p-0.100_t-1.300_i_c_l_B-0.500_i_l_B-1.000_0.csv",
            "iter-100_l-50_h-50_d_A-0.250_d_B-0.250_m_r_A-2_m_r_B-2_r_m_A-0.200_r_r_A-0.500_g_r_A-0.200_d_r_A-0.500_t_l_A-0.900_p-0.100_t-1.300_i_c_l_B-0.500_i_l_B-1.000_0.csv",
            "iter-100_l-50_h-50_d_A-0.250_d_B-0.250_m_r_A-2_m_r_B-2_r_m_A-0.200_r_r_A-0.500_g_r_A-0.200_d_r_A-0.200_t_l_A-0.900_p-0.100_t-1.300_i_c_l_B-0.500_i_l_B-1.000_0.csv",
            "iter-100_l-50_h-50_d_A-0.250_d_B-0.250_m_r_A-2_m_r_B-2_r_m_A-0.200_r_r_A-0.500_g_r_A-0.100_d_r_A-0.200_t_l_A-0.900_p-0.100_t-1.300_i_c_l_B-0.500_i_l_B-1.000_0.csv",
            "iter-100_l-50_h-50_d_A-0.250_d_B-0.250_m_r_A-2_m_r_B-2_r_m_A-0.200_r_r_A-0.300_g_r_A-0.500_d_r_A-0.200_t_l_A-0.900_p-0.100_t-1.300_i_c_l_B-0.500_i_l_B-1.000_0.csv",
            "iter-100_l-50_h-50_d_A-0.250_d_B-0.250_m_r_A-2_m_r_B-2_r_m_A-0.200_r_r_A-0.300_g_r_A-0.500_d_r_A-0.500_t_l_A-0.900_p-0.100_t-1.300_i_c_l_B-0.500_i_l_B-1.000_0.csv",
            "iter-100_l-50_h-50_d_A-0.250_d_B-0.250_m_r_A-2_m_r_B-2_r_m_A-0.200_r_r_A-0.200_g_r_A-0.500_d_r_A-0.200_t_l_A-0.900_p-0.100_t-1.300_i_c_l_B-0.500_i_l_B-1.000_0.csv",
            "iter-100_l-50_h-50_d_A-0.250_d_B-0.250_m_r_A-2_m_r_B-2_r_m_A-0.200_r_r_A-0.200_g_r_A-0.200_d_r_A-0.100_t_l_A-0.900_p-0.100_t-1.300_i_c_l_B-0.500_i_l_B-1.000_0.csv"
           ]

MeanCoop_Duration_s = []
Attack_Duration_s = []

for ids, s in enumerate(scenario):
    print(s)
    try:
        file = str(s).split("del")[-1]
        file = str(s).split(".csv")[0]

        i_c_l_B = float(file.split("i_c_l_B-")[-1][:5])
        threshold_life_A = float(file.split("t_l_A-")[-1][:5])
        
        Allmoves_A = pd.read_csv("Results_AB/allmoves_A/" + str(s))
        Allmoves_A = Allmoves_A.replace(-1, 0)
        
        Allmoves_B = pd.read_csv("Results_AB/allmoves_B/" + str(s))
        
        Attack_Duration = []
        MeanCoop_Duration = []
        for idx, x in enumerate(Allmoves_A["simul_step"]):
            if Allmoves_A["action_type"][idx] == "DC":
                LifeD = Allmoves_A[(Allmoves_A["action_type"] != "D") & (Allmoves_A["action_type"] != "R")]
                LifeD = LifeD[(LifeD["action_type"] == "DC") & (LifeD["x location"] == LifeD["x location"][idx])
                    & (LifeD["y location"] == LifeD["y location"][idx])
                    | (LifeD["action_type"] == "M") & (LifeD["x destination"] == LifeD["x location"][idx]) 
                    & (LifeD["y destination"] == LifeD["y location"][idx])]
                TimestampL = LifeD["simul_step"].tolist()
                LifeA = LifeD["action_type"].tolist()

                if LifeA[0] == "DC":
                    LifeA.insert(0, "M")
                    TimestampL.insert(0, 0)
                EndList = len(LifeA) - 1 - LifeA[::-1].index('DC')
                TimestampL = TimestampL[0:EndList+1]
                LifeA = LifeA[0:EndList+1]
                # print(TimestampL)
                # print(LifeA)
                
                k = 0      
                Debut = []
                for idt, t in enumerate(TimestampL):
                    if LifeA[idt] == "DC":
                        if idt < len(TimestampL)-1:
                            Debut.append(TimestampL[idt+1])
                        if k == 0:
                            Duration1 = t - TimestampL[0]
                            Attack_Duration.append(Duration1)
                            CoopInter = Allmoves_B[(Allmoves_B["simul_step"] >= TimestampL[0]-10) & (Allmoves_B["simul_step"] <= t+10)]
                            MeanCoop = np.sum(CoopInter["cooperators"]/ (CoopInter["cooperators"]+CoopInter["defectors"])).tolist()/len(CoopInter)
                            MeanCoop_Duration.append(MeanCoop)
                            k+=1
                        elif k > 0:
                            Duration2 = t - Debut[k-1]
                            Attack_Duration.append(Duration2)
                            CoopInter2 = Allmoves_B[(Allmoves_B["simul_step"] >= Debut[k-1]-10) & (Allmoves_B["simul_step"] <= t+10)]
                            MeanCoop2 = np.sum(CoopInter2["cooperators"]/ (CoopInter2["cooperators"]+CoopInter2["defectors"])).tolist()/len(CoopInter2)
                            MeanCoop_Duration.append(MeanCoop2)
                            k += 1
        # fig = plt.figure(figsize = (15,15))
        # plt.scatter(MeanCoop_Duration, Attack_Duration)
        
        fig, ax = plt.subplots(1,1,figsize=(15, 15))
        
        n, bins, patches = ax.hist(MeanCoop_Duration, 50, density=True, facecolor='b', alpha=1.0, weights = Attack_Duration)
        ax.set_xlabel(r'Average cooperation level, $\overline{C_L}$', fontsize=24)
        ax.set_ylabel('Probability Density', fontsize=24)
        ax.set_xlim(0.45,1.05)
        
        ax.tick_params(labelsize=20)
    
    except:
        print("Exception")
        pass
    MeanCoop_Duration_s.append(MeanCoop_Duration)
    Attack_Duration_s.append(Attack_Duration)
    
    with open("MeanCoop_Duration"+str(ids)+".txt", 'w') as f:
        for s in MeanCoop_Duration:
            f.write(str(s) + '\n')
            
    with open("Attack_Duration"+str(ids)+".txt", 'w') as f:
        for s in Attack_Duration:
            f.write(str(s) + '\n')
```
